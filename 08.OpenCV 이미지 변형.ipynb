```
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploads = files.upload()

img = cv2.imread("cat.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.show
```

cv2.resize(image, dsize, fx, fy, interpolation): 이미지의 크기를 조절합니다.
1.   dsize: Manual Size(수동사이즈)
2.   fx: 가로 비율
3.   fy: 세로 비율
4.   interpolation: 보간법

보간법의 종류
1.   INTER_CUBIC: 사이즈를 크게 할 때 주로 사용합니다.
2.   INTER_AREA: 사이즈를 작게 할 때 주로 사용합니다.

보간법은 사이즈가 변할 때 픽셀 사이의 값을 조절하는 방법을 의미합니다.

```
# 확대
expand = cv2.resize(img,None,fx=2.0,fy=2.0,interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand,cv2.COLOR_BGR2RGB))
plt.show
```

```
# 축소
shrink = cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink,cv2.COLOR_RGB2BGR))
plt.show
```

```
# 인덱스 BGR 제외하고 resize 됨.
print(img.shape) # (380, 441, 3)
print(expand.shape) # (760, 882, 3)
print(shrink.shape) # (190, 220, 3)
```

cv2.warpAffine(image, M, dsize): 이미지의 위치를 변경합니다.
*   M: 변환 행렬
*   dsize: Manual Size

이동 변환행렬 : M = np.float32([[1,0,x축 이동],[0,1,y축 이동]])

회전 변환행렬 : M = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5) -> center, angle, scale

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("cat.jpg")
height,width = img.shape[:2]

#변환행렬 설정
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show
```

```
img = cv2.imread("cat.jpg")
height, width = img.shape[:2]

# center,angle,scale
M = cv2.getRotationMatrix2D((width/2,height/2),90,0.5)
dst = cv2.warpAffine(img,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show
```

```
#마지막 혼자 테스트

img_ = cv2.imread("cat.jpg")
#print(img_.shape)
roi = img_[100:250,100:250]
#print(roi)
img_[:150,:150] = roi

height,width = img_.shape[:2]
print(height)
print(width)
M = np.float32([[1,0,width/8],[0,1,height/15]])

dst = cv2.warpAffine(img_,M,(width,height))

height_, width_ = dst.shape[:2]
M1 = cv2.getRotationMatrix2D((width/2,height/2),120,0.5)
dst_ = cv2.warpAffine(dst,M1,(width,height))

plt.imshow(dst_)
plt.show
```
