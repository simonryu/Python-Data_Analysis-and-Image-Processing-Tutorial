cv2.resize(image, dsize, fx, fy, interpolation):   .
1.   dsize: Manual Size()
2.   fx:  
3.   fy:  
4.   interpolation: 

 
1.   INTER_CUBIC:      .
2.   INTER_AREA:      .

         .

```
# 
expand = cv2.resize(img,None,fx=2.0,fy=2.0,interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand,cv2.COLOR_BGR2RGB))
plt.show
```

```
# 
shrink = cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink,cv2.COLOR_RGB2BGR))
plt.show
```

cv2.warpAffine(image, M, dsize):   .
*   M:  
*   dsize: Manual Size

  : M = np.float32([[1,0,x ],[0,1,y ]])

  : M = cv2.getRotationMatrix2D((width / 2, height / 2), 90, 0.5) -> center, angle, scale

```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread("cat.jpg")
height,width = img.shape[:2]

# 
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show
```

```
img = cv2.imread("cat.jpg")
height, width = img.shape[:2]

# center,angle,scale
M = cv2.getRotationMatrix2D((width/2,height/2),90,0.5)
dst = cv2.warpAffine(img,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show
```
