```
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_1 = cv2.imread("image_1.jpg")
img_2 = cv2.imread("image_2.png")

plt.imshow(cv2.cvtColor(img_1,cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(img_2,cv2.COLOR_BGR2RGB))
plt.show
```


cv2.add(): Saturation 연산을 수행합니다.
*   Saturation: 0보다 작으면 0, 255보다 크면 255로 표현합니다.

np.add(): Modulo 연산을 수행합니다. (modulo = 나머지)
*   Modulo: 256은 0, 257은 1로 표현합니다.




```
# Modulo 연산
add_img1 = np.add(img_1 ,img_2)
plt.imshow(cv2.cvtColor(add_img1,cv2.COLOR_BGR2RGB))
plt.show
```

```
# Saturation 연산
add_img2 = cv2.add(img_1, img_2)
plt.imshow(cv2.cvtColor(add_img2,cv2.COLOR_BGR2RGB))
plt.show
```
