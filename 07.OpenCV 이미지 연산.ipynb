```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("cat.jpg")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.show
```

```
#import numpy as np
#print(np.size(img))
#print(np.shape(img))

# 픽셀 수
print(img.size)
# 이미지 크기
print(img.shape)
# 특정 픽셀의 값 (B,G,R)
px = img[379, 440]
print(px)
print(px[0])
print(px[1])
print(px[2]) 
```

```
#특정필셀 변경하기 (Slice 사용)))
img[:100, :100] = [255,255,255]
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.show
```

```
# ROI 활용 (특정부분만 가져옴)
roi = img[200:300, 200:300]
img[:100,:100] = roi
plt.imshow(cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
plt.show
```
