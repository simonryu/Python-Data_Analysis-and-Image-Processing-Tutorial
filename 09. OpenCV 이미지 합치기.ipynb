cv2.add(): Saturation 연산을 수행합니다.
*   Saturation: 0보다 작으면 0, 255보다 크면 255로 표현합니다.

np.add(): Modulo 연산을 수행합니다.
*   Modulo: 256은 0, 257은 1로 표현합니다.

```
import numpy as np

img1 = cv2.imread("image_1.jpg",cv2.IMREAD_COLOR)
img2 = cv2.imread("image_2.png",cv2.IMREAD_COLOR)

img_a1 = cv2.add(img1,img2)
plt.figure(1)
plt.imshow(cv2.cvtColor(img_a1,cv2.COLOR_BGR2RGB))
plt.show

img_a2 = np.add(img1,img2)
plt.figure(2)
plt.imshow(cv2.cvtColor(img_a2, cv2.COLOR_BGR2RGB))
plt.show

```
