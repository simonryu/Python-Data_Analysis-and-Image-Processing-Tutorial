```
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("gray_image.jpg")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

*     (Blurring = Smoothing)   .
*         .
*   (Kernel)       
.

```
size = 4
kernel = np.ones((size,size),np.float32) / (size**2)
print(kernel)

#   
dst = cv2.filter2D(image,-1,kernel)
plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show()
```

**Basic Blurring**

```
dst_blur = cv2.blur(image,(4,4))
plt.imshow(cv2.cvtColor(dst_blur,cv2.COLOR_BGR2RGB))
plt.show()
```

**Gaussian Blur**

```
# kernel_size: 
dst_GB = cv2.GaussianBlur(image,(7,7),0)
plt.imshow(cv2.cvtColor(dst_GB,cv2.COLOR_BGR2RGB))
plt.show()
```
