```
import matplotlib.pyplot as plt
import cv2

basic_img = cv2.imread("cat.jpg",cv2.IMREAD_COLOR)
plt.imshow(cv2.cvtColor(basic_img,cv2.COLOR_BGR2RGB))
plt.show
```

```
gray_img = cv2.imread("cat.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(gray_img,cv2.COLOR_GRAY2RGB))
plt.show
```

```
#  np img  color cvt
gray2_img = cv2.cvtColor(basic_img,cv2.COLOR_BGR2GRAY)
#cvt plt BGR   cvt
plt.imshow(cv2.cvtColor(gray2_img,cv2.COLOR_GRAY2BGR))
plt.show
```
